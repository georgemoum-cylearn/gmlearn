```
#εισαγωγή βιβλιοθηκών - πακέτων
import pycaret
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pycaret.classification import *
```

```
#εισαγωγή του συνόλου δεδομένων
url = 'https://raw.githubusercontent.com/kdemertzis/EKPA/main/Data/dga_domain.csv'
df = pd.read_csv(url)
#έλεγχος σχήματος συνόλου δεδομένων
print(df.shape)
```

```
#ανακάτεμα των δεδομένων
df = df.sample(frac = 1)
print(df.head())
```

```
from pycaret.classification import setup, compare_models

# Εκτύπωση των ονομάτων των στηλών για να ελέγξετε την ακριβή ονομασία της στήλης 'class'
print(df.columns)

# Αφαίρεση κενών από τα ονόματα των στηλών
df.columns = df.columns.str.strip()

# Συνεχίστε με την ρύθμιση του PyCaret
clf = setup(data = df, target = 'class', session_id = 123)
best = compare_models(cross_validation=False, n_select = 1)
%load_ext tensorboard
```

```
#αναφορά των υπερπαραμέτρων του μοντέλου με την υψηλότερη απόδοση
try:
    print(best)
except NameError:
    print("Error: The variable 'best' is not defined. Please ensure all previous cells were run successfully.")
```

```
#μέτρα αξιολόγησης του μοντέλου με την υψηλότερη απόδοση
evaluate_model(best)
```
